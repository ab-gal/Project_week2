### Project Scoping:

Explenation: Scoping defines the project’s boundaries, specifying objectives, required resources, key features, priorities, and constraints. It ensures team alignment and resource availability. By clear scoping, projects are more manageable and focused, reducing the risk of scope creep.

### Roadmapping:

Explenation: A roadmap lays out the project’s timeline and major milestones, from data collection to model deployment. It helps in setting expectations, managing stakeholders, and tracking progress. In data analytics, roadmaps might include stages such as data preprocessing, analysis, modeling, and reporting.
